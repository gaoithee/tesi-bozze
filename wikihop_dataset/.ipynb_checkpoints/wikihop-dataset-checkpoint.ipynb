```python
df = pd.read_csv('wikihop-test.csv')
df.head()
```

```python
df['query'][:5]
```

```python
# Funzione di preprocessing
def preprocess_question(text):
    # Usa una regex per trovare la parte tra virgolette che rappresenta la domanda
    match = re.search(r'"(.*?)"', text)
    if match:
        return match.group(1)
    return text  # Restituisce il testo originale se non trova corrispondenza

# Applica la funzione di preprocessing alla colonna 'query'
df['cleaned_query'] = df['query'].apply(preprocess_question)
df.head()
```

```python
df = df.drop('Unnamed: 0', axis=1)
df.head()
```
